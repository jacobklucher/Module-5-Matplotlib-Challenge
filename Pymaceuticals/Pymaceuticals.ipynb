### Analysis

- Ramicane and Capomulin had the lowest tumor volume values for mean, median, variance, standard deviation, and standard error. These two treatments also had more observed timepoints then any other treatment.
- Infubinol is the only treatment calculated to have a potential outlier.
- From the line chart of Capomulin treatment, it appears that the tumor volume increases until the 20th day of treatment then after that there is a drop off in tumor volume.
- From the scatter plot, we can see the correlation between mouse weight and the average tumor volume is 0.84, which is a fairly high correlation. This means that mouse weight probably correlates to average tumor volume.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Combine the data into a single DataFrame
inner_merge_df = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="inner")

# Display the data table for preview
inner_merge_df.head()
```

```python
# Checking the number of mice.
len(inner_merge_df["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
id_duplicate = inner_merge_df[["Mouse ID", "Timepoint"]].duplicated(keep=False)
duplicate_mouse_ids = inner_merge_df.loc[id_duplicate, "Mouse ID"].unique()
duplicate_mouse_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_mouse_rows = inner_merge_df.loc[inner_merge_df["Mouse ID"] == "g989"]
duplicate_mouse_rows
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = inner_merge_df[inner_merge_df["Mouse ID"] != "g989"]
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_df["Mouse ID"].unique())
```

## Summary Statistics

```python
tumor_regimen = clean_df.groupby("Drug Regimen")
volume = tumor_regimen["Tumor Volume (mm3)"]
mean = volume.mean()
median = volume.median()
variance = volume.var()
stddev = volume.std()
```

```python
samples = volume.sample(100)
samples
```

```python
sems = volume.apply(st.sem)
sems
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.
tumor_summary = pd.DataFrame({
                "Mean Tumor Volume": mean,
                "Median Tumor Volume": median,
                "Tumor Volume Variance": variance,
                "Tumor Volume Std. Dev.": stddev,
                "Tumor Volume Std. Err.": sems
})

tumor_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
tumor_regimen_agg = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(
     mean = "mean",
     median = "median",
     var = "var",
     std = "std",
     sem = "sem"
 )
tumor_regimen_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_volume_df = clean_df.groupby("Mouse ID")["Timepoint"].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume = pd.merge(final_volume_df, clean_df, on=["Mouse ID", "Timepoint"], how="left")
final_tumor_volume[["Mouse ID", "Timepoint", "Tumor Volume (mm3)", "Drug Regimen"]].head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
for treatment in treatments:

    # Locate the rows which contain mice on each drug and get the tumor volumes
    subset = final_tumor_volume[final_tumor_volume["Drug Regimen"] == treatment]

    tumor_vol_data.append(subset["Tumor Volume (mm3)"].values)

    Q1 = subset["Tumor Volume (mm3)"].quantile(0.25)
    Q3 = subset["Tumor Volume (mm3)"].quantile(0.75)
    IQR = Q3 - Q1

    # Determine outliers using upper and lower bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = subset[(subset["Tumor Volume (mm3)"] < lower_bound) | (subset["Tumor Volume (mm3)"] > upper_bound)]

    print(f"{treatment}'s potential outliers: ")
    print(f"{outliers["Tumor Volume (mm3)"]}\n")
```

```python
zipped_data = zip(treatments, tumor_vol_data)
tumor_vol_dict = {treatment: vol for treatment, vol in zipped_data}
```

```python
tumor_vol_df = pd.DataFrame(tumor_vol_dict)
tumor_vol_df
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
tumor_vol_df.boxplot(grid=False)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Make sure mouse exists
test_mouse = clean_df[(clean_df["Mouse ID"] == "l509")]
test_mouse
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
mouse = "l509"
treatment = "Capomulin"

single_mouse_tumor_df = clean_df[(clean_df["Mouse ID"] == mouse) & (clean_df["Drug Regimen"] == treatment)]

timepoints = single_mouse_tumor_df["Timepoint"]
tumor_volumes = single_mouse_tumor_df["Tumor Volume (mm3)"]

plt.plot(timepoints, tumor_volumes, color="blue")
plt.title(f"{treatment} treatment of mouse {mouse}")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_regimen = clean_df[clean_df["Drug Regimen"] == "Capomulin"]

avg_tumor_volume = capomulin_regimen.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()

# Get the first mouse weight for each id
mouse_weights = capomulin_regimen.groupby("Mouse ID")["Weight (g)"].first()

weight_volume_data = pd.DataFrame({"Mouse Weight (g)": mouse_weights, "Average Tumor Volume (mm3)": avg_tumor_volume})

plt.scatter(weight_volume_data["Mouse Weight (g)"], weight_volume_data["Average Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = weight_volume_data["Mouse Weight (g)"]
y_values = weight_volume_data["Average Tumor Volume (mm3)"]

corr_coef = x_values.corr(y_values)
print(f"The correlation between mouse weight and the average tumor volume is {round(corr_coef, 2)}")

slope, intercept, r_value, p_value, std_err = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept

plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()
```
